```python
#Import libraries
import pandas as pd
import nltk
import re
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
data=[]
```

```python
#Read the Dataset
df=pd.read_csv('IMDB Dataset.csv')
```

```python
df
```

```python
#Text Preprocessing

for i in range(0,1000):
  review=df['review'][i]

  #remove uneccessary things
  review=re.sub('[^a-zA-Z]',' ',review )

  #Lower case text
  review=review.lower()

  #split the text
  review=review.split()

  #Stemming & Remove Stopwords
  review=[ps.stem(word) for word in review if not word in 
          set(stopwords.words('english'))]

  review=' '.join(review)
  data.append(review)
```

```python
#Tokenization or Vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=3000)
```

```python
cv
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
#Splitting the data into x & y

x=cv.fit_transform(data).toarray()
y=df.iloc[:1000,1:2].values

ct=ColumnTransformer([("one", OneHotEncoder(), [0])], remainder="passthrough")
y=ct.fit_transform(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
import numpy as np
```

```python
x=np.asarray(x).astype(np.float32)
print(x)
```

```python
y=y[:,1:2]
y = np.asarray(y).astype(np.float32)
print(y)
```

```python
#Split the  data into train & test

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(1565,activation='relu'))
model.add(Dense(3000,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=10)
```

```python
#save the model
model.save('movie_review.h5')
```
